# College Majors

## Introduction to the College Majors Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.shape[0]
print("Before cleaning:", raw_data_count)
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print("After cleaning:", cleaned_data_count)
```

## Scatter Plots

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter", title="Sample_size and Median")
```

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter", title="Sample_size and Unemployment_rate")
```

```python
recent_grads.plot(x="Full_time", y="Median", kind="scatter", title="Full_time and Median")
```

```python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter", title="ShareWomen and Unemployment_rate")
```

```python
recent_grads.plot(x="Men", y="Median", kind="scatter", title="Men and Median")
```

```python
recent_grads.plot(x="Women", y="Median", kind="scatter", title="Women and Median")
```

## Histograms

```python
columns = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(10,35))
for i in range(8):
    ax = fig.add_subplot(8,1,i+1)
    ax = recent_grads[columns[i]].hist(bins=20)
    ax.set_title(columns[i])
```

## Scatter Matrix Plots

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(recent_grads[["Sample_size", "Median"]])
```

```python
scatter_matrix(recent_grads[["Sample_size", "Median", "Unemployment_rate"]], figsize=(10,10))
```
